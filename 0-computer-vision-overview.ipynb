# AI

Ở cấp độ cơ bản nhất, lĩnh vực trí tuệ nhân tạo sử dụng khoa học máy tính và dữ liệu để cho phép giải quyết vấn đề trong máy móc.   
Machine Learning (ML) là tập con của AI, đề cập đến việc nghiên cứu các hệ thống máy tính có khả năng học hỏi và thích ứng tự động từ kinh nghiệm mà không cần được lập trình rõ ràng.  
Deep Learning là một tập con của ML là một kỹ thuật học máy xếp các thuật toán và đơn vị tính toán— hoặc tế bào thần kinh — vào cái được gọi là mạng lưới thần kinh nhân tạo

Một cách dễ hiểu, AI là một hệ thống máy tính tiếp nhận các dữ liệu đầu vào và xử lý, biến đổi nó và đưa ra các đầu ra phù hợp:



# Computer Vision

Thị giác máy tính là một công nghệ mà máy sử dụng để tự động nhận biết và mô tả hình ảnh một cách chính xác và hiệu quả. Ngày nay, các hệ thống máy tính có quyền truy cập vào khối lượng lớn hình ảnh và dữ liệu video bắt nguồn từ hoặc được tạo bằng điện thoại thông minh, camera giao thông, hệ thống bảo mật và các thiết bị khác. Ứng dụng thị giác máy tính sử dụng trí tuệ nhân tạo và máy học (AI/ML) để xử lý những dữ liệu này một cách chính xác nhằm xác định đối tượng và nhận diện khuôn mặt, cũng như phân loại, đề xuất, giám sát và phát hiện.  



Các thành phần của thị giác máy tính bao gồm:
- Chuyển đổi và xử lý tín hiệu thu được từ cảm biến hình ảnh (camera) thành dữ liệu ảnh trong lập trình
- Các thuật toán định dạng và nâng cao chất lượng hình ảnh
- Lưu trữ và truyền tải hình ảnh
- Nhận dạng hình ảnh sử dụng trí tuệ nhân tạo

Trong AI, Computer Vision là thuật ngữ để chỉ các thuật toán, mô hình chuyên sử dụng để xử lý hình ảnh.  
Trong Computer Vision, AI được sử dụng là khối xử lý & nhận dạng hình ảnh
