## Python là ngôn ngữ thông dịch
Giống như bash, shell, bat, Python là một ngôn ngữ thông dịch:  

  

Thông dịch là quá trình chuyển đổi từ mã nguồn thành mã máy ngay tại câu lệnh hiện tại.    
Quá trình thực thi mã nguồn bao gồm biên dịch mã nguồn thành mã trung gian được gọi là bytecode, sau đó bytecode được trình thông dịch đọc và thực thi ở máy ảo PVM

Mở Terminal và trải nghiệm Python dưới dạng một shell:  

````python
!python
````

Python shell sẽ hiện ra và ngay lập tức có thể lập trình  
Hãy nhập dòng lệnh đầu tiên của bạn  


````python
print('hello work')
````

Lệnh được thực thi ngay lập tức, đó chính là đặc điểm của thông dịch.

## Chương trình Python đầu tiên

Chương trình Python thường được viết dưới dạng các script, thay vì sử dụng như một shell  
Tạo một file có tên ```helloworld.py``` và đưa code sau đây vào:

````python
# -*- coding: utf-8 -*-
# @Author        : Cuong Tran

import os
import sys

def main():
    print('Hello worlds from ', os.name)

if __name__ == '__main__':
    main()
    input()
````

Như đã trình bày ở phần trên, Python là một ngôn ngữ thông dịch điển hình, nó sẽ đưa mã nguồn về dạng bytecode và thực thi bytecode bằng PVM  
Quá trình khởi chạy mã nguồn Python như sau:

````python
# Biên dịch mã nguồn thành bytecode
!python -m py_compile helloworlds.py

# Thư mục __pycache__ sẽ được tạo và nó chứa bytecode sau khi biên dịch
%ls -l __pycache__

# Khởi chạy bytecode
!python ./__pycache__/helloworlds.cpython-310.pyc
````

Python hỗ trợ rút ngắn quá trình trên chỉ bằng một câu lệnh duy nhất:

````python
!python helloworlds.py
````

Từ đây về sau, chúng ta chỉ sử dụng phương pháp rút ngắn !!
